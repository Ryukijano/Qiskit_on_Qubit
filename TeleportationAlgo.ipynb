```python
#Quantum Teleportation Algorithm
#importing qiskit to workspace
from qiskit import *
```

```python
#building a quantum circuit
#utilizing 3 C-bits and 3 Q-bits
circuit = QuantumCircuit(3,3)
```

```python
%matplotlib inline
circuit.draw(output = 'mpl')
```

```python
#Ambition: Teleporting Qo to Q2, Q1 will be a resource in order to do this process.
#In simpler terms, we are gonna transfer the state of Quantum bit zero to Quantum bit two.
circuit.x(0)
circuit.barrier()
circuit.draw(output = 'mpl')
```

```python
#creating entanglement between Q1 and Q2
circuit.h(1)
circuit.cx(1,2)
circuit.draw(output = 'mpl')
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
circuit.measure([0,1], [0,1])
circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output = 'mpl')
```

```python
#Teleportation Successfully Accomplished!
#Final Step: verification of the algorithm

circuit.measure(2,2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
